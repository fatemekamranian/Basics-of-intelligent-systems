```python
import numpy as np
import matplotlib.pyplot as plt

# Define McCulloch-Pitts Neuron class with ReLU activation
class McCulloch_Pitts_neuron():
    def __init__(self, weights, threshold):
        self.weights = weights    # Define weights
        self.threshold = threshold    # Define threshold

    def model(self , x):
      #define model with threshold
      if self.weights @ x >= self.threshold:
        return 1
      else:
        return 0

# Define the function to classify the area inside the triangle using McCulloch-Pitts neurons
def Area(x, y):
    # Coordinates of the triangle vertices A(2, 2), B(3, 0), C(1, 0)
    A = np.array([2, 2])
    B = np.array([3, 0])
    C = np.array([1, 0])

    # Vectors representing the edges of the triangle
    AB = B - A
    BC = C - B
    CA = A - C

    # Perpendicular vectors (normal vectors to the triangle edges)
    normal_AB = np.array([AB[1], -AB[0]])  # Perpendicular to AB
    normal_BC = np.array([BC[1], -BC[0]])  # Perpendicular to BC
    normal_CA = np.array([CA[1], -CA[0]])  # Perpendicular to CA

    # Define McCulloch-Pitts neurons for each side of the triangle with ReLU activation
    neur1 = McCulloch_Pitts_neuron(normal_AB, 0)  # Neuron for edge AB
    neur2 = McCulloch_Pitts_neuron(normal_BC, 0)  # Neuron for edge BC
    neur3 = McCulloch_Pitts_neuron(normal_CA, 0)  # Neuron for edge CA

    # Use the neurons to classify if the point is inside or outside the triangle
    z1 = neur1.model(np.array([x - A[0], y - A[1]]))  # Point relative to AB
    z2 = neur2.model(np.array([x - B[0], y - B[1]]))  # Point relative to BC
    z3 = neur3.model(np.array([x - C[0], y - C[1]]))  # Point relative to CA

    # If the point is on the correct side for all edges, it's inside the triangle
    if z1 > 0 and z2 > 0 and z3 > 0:
        return [1]  # Inside the triangle (green)
    else:
        return [0]  # Outside the triangle (red)

# Generate random data points
num_points = 2000
x_values = np.random.uniform(0, 4, num_points)  # Updated x-axis limits
y_values = np.random.uniform(-1, 3, num_points)  # Updated y-axis limits

# Initialize lists to store data points for different z5 values
red_points = []
green_points = []

# Evaluate data points using the Area function
for i in range(num_points):
    z5_value = Area(x_values[i], y_values[i])
    if z5_value == [0]:  # z5 value is 0 (outside)
        red_points.append((x_values[i], y_values[i]))
    else:  # z5 value is 1 (inside)
        green_points.append((x_values[i], y_values[i]))

# Separate x and y values for red and green points
red_x, red_y = zip(*red_points)
green_x, green_y = zip(*green_points)

# Plotting
plt.figure(figsize=(8, 6))
plt.scatter(red_x, red_y, color='red', label='Outside the triangle')
plt.scatter(green_x, green_y, color='green', label='Inside the triangle')

# Plotting the triangle using the vertices
triangle_x = [2, 3, 1, 2]
triangle_y = [2, 0, 0, 2]
plt.plot(triangle_x, triangle_y, color='black', linestyle='-', label='Triangle Boundary')

plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('McCulloch-Pitts Neuron Classifier for Triangle Area')

plt.grid(True)

# Set axis limits
plt.xlim(0, 4)
plt.ylim(-1, 3)

# Position the legends at the top and right
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1.0))

plt.show()
```

# Relu Activation Function

# Sigmoid Activation Function

```python
import numpy as np
import matplotlib.pyplot as plt

# Define McCulloch-Pitts Neuron class with ReLU activation
class McCulloch_Pitts_neuron():
    def __init__(self, weights, threshold):
        self.weights = weights    # Define weights
        self.threshold = threshold    # Define threshold

    def sigmoid(self, x):
        # Apply Sigmoid activation function
        return 1 / (1 + np.exp(-x))

    def model(self, x):
        return self.sigmoid(self.weights @ x)

# Define the function to classify the area inside the triangle using McCulloch-Pitts neurons
def Area(x, y):
    # Coordinates of the triangle vertices A(2, 2), B(3, 0), C(1, 0)
    A = np.array([2, 2])
    B = np.array([3, 0])
    C = np.array([1, 0])

    # Vectors representing the edges of the triangle
    AB = B - A
    BC = C - B
    CA = A - C

    # Perpendicular vectors (normal vectors to the triangle edges)
    normal_AB = np.array([AB[1], -AB[0]])  # Perpendicular to AB
    normal_BC = np.array([BC[1], -BC[0]])  # Perpendicular to BC
    normal_CA = np.array([CA[1], -CA[0]])  # Perpendicular to CA

    # Define McCulloch-Pitts neurons for each side of the triangle with ReLU activation
    neur1 = McCulloch_Pitts_neuron(normal_AB, 0)  # Neuron for edge AB
    neur2 = McCulloch_Pitts_neuron(normal_BC, 0)  # Neuron for edge BC
    neur3 = McCulloch_Pitts_neuron(normal_CA, 0)  # Neuron for edge CA

    # Use the neurons to classify if the point is inside or outside the triangle
    z1 = neur1.model(np.array([x - A[0], y - A[1]]))  # Point relative to AB
    z2 = neur2.model(np.array([x - B[0], y - B[1]]))  # Point relative to BC
    z3 = neur3.model(np.array([x - C[0], y - C[1]]))  # Point relative to CA

    # If the point is on the correct side for all edges, it's inside the triangle
    if z1 > 0.5 and z2 > 0.5 and z3 > 0.5:
      return [1]  # Inside the triangle (green)
    else:
      return [0]  # Outside the triangle (red)

# Generate random data points
num_points = 2000
x_values = np.random.uniform(0, 4, num_points)  # Updated x-axis limits
y_values = np.random.uniform(-1, 3, num_points)  # Updated y-axis limits

# Initialize lists to store data points for different z5 values
red_points = []
green_points = []

# Evaluate data points using the Area function
for i in range(num_points):
    z5_value = Area(x_values[i], y_values[i])
    if z5_value == [0]:  # z5 value is 0 (outside)
        red_points.append((x_values[i], y_values[i]))
    else:  # z5 value is 1 (inside)
        green_points.append((x_values[i], y_values[i]))

# Separate x and y values for red and green points
red_x, red_y = zip(*red_points)
green_x, green_y = zip(*green_points)

# Plotting
plt.figure(figsize=(8, 6))
plt.scatter(red_x, red_y, color='red', label='Outside the triangle')
plt.scatter(green_x, green_y, color='green', label='Inside the triangle')

# Plotting the triangle using the vertices
triangle_x = [2, 3, 1, 2]
triangle_y = [2, 0, 0, 2]
plt.plot(triangle_x, triangle_y, color='black', linestyle='-', label='Triangle Boundary')

plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Sigmoid')

plt.grid(True)

# Set axis limits
plt.xlim(0, 4)
plt.ylim(-1, 3)

# Position the legends at the top and right
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1.0))

plt.show()
```
